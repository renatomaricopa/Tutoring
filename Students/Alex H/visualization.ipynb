```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("HR_Baseline_Data.csv")
pd.set_option('display.max_columns', None)
df
```

```python
corr_satisfaction = df['TrainingTimesLastYear'].corr(df['JobSatisfaction'])     # correlation
corr_satisfaction
```

```python
# Calculate a linear regression line;  capture the following results
d1_const = 0     # regression line constant
x1_coef = 0     # x coefficient
lm = LinearRegression()
x = np.array(df['TrainingTimesLastYear']).reshape(-1,1)
y = np.array(df['JobSatisfaction'])
lm.fit(x, y)
Yhat = lm.predict(x)
d1_const = lm.intercept_
x1_coef = lm.coef_[0]
```

```python
# Plot it!
# plot x and y as a scatter plot  (dots should be blue)
# On the same chart, plot the regression line without marker (line should be black)
# Choose any style you desire, include x and y axis labels and a title that includes the file name
# Annotate with a box in the upper right corner of the grid as described above
x = np.array(df['TrainingTimesLastYear'])
fig, ax = plt.subplots()
plt.scatter(x, y, 'o')
plt.plot(x, x1_coef* x + d1_const, color ='black')
plt.title('Training Times Last Year vs. Job Satisfaction')
plt.xlabel("Training Times", fontsize = 10)
plt.ylabel("Job Satisfaction", fontsize = 10)

```
